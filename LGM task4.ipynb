## Task 4 - Image to Pencil Sketch with Python:
## Level - Beginner

#### Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

#### Reading the image

```python
image = cv2.imread('Eiffel-Tower.webp')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), cmap = 'gray')
plt.show()
```

#### Converting image to gray-scale.

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap = 'gray')
plt.show()
```

#### Inverting the gray-scale image(Negative)

```python
inverted_img = 255 - gray_img
```

```python
plt.imshow(inverted_img, cmap = 'gray')
plt.show()
```

#### Blurring the inverted gray-scale image.

```python
blur_img = cv2.GaussianBlur(inverted_img, (21,21),0)
plt.imshow(blur_img, cmap = 'gray')
plt.show()
```

#### Pencil Steck

```python
inverted_blur = 255 - blur_img
pencilsketch = cv2.divide(gray_img, inverted_blur, scale = 256.0)
plt.imshow(pencilsketch, cmap = 'gray')
plt.show()
```

#### Orignal image / Pencil sketch

```python
# Original Image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), cmap = 'gray')
plt.show()

# Pencil Sketch 
plt.imshow(pencilsketch, cmap = 'gray')
plt.show()
```
